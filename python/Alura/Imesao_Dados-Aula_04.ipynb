```python
#desafio 01
contagem_moa_top = dados_resultados.select_dtypes('int64').drop('n_moa', axis = 1).sum().sort_values(ascending=False).index[:10]
contagem_moa_top
top = []
for i in contagem_moa_top:
    if i[len(i)-1] == 'r':
        top.append('inhibitor')
    elif i[len(i)-1] == 't': 
        if i[len(i)-8] == '_':
            top.append('agonist')
        else:
            top.append('antagonist')
    else:
        top.append(i)
top
```

```python
#Desafio 02
eh_controle=[]
for i in dados['tratamento']:
    if i[len(i)-1] == 'e':
        eh_controle.append(1)
    else:
        eh_controle.append(0)
dados_combinados['eh_controle'] = eh_controle

```

```python
###Desafio 03: Criar três colunas para indicar o tempo 24, 48, 72
t24 = []
t48 = []
t72 = []
for i in dados['tempo']:
    if i == 24:
        t24.append(1)
        t48.append(0)
        t72.append(0)
    elif i == 48:
        t24.append(0)
        t48.append(1)
        t72.append(0)
    else:
        t24.append(0)
        t48.append(0)
        t72.append(1) 
dados_combinados['Tempo = 24'] = t24
dados_combinados['Tempo = 48'] = t48
dados_combinados['Tempo = 72'] = t72
```

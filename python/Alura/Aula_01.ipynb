```python
import matplotlib.pyplot as plot

tipos_de_tratamento = dados.tratamento.value_counts()
labels = dados.tratamento.unique()

plot.pie(tipos_de_tratamento, labels=labels, autopct='%1.1f%%')
plot.title('Tipos de tratamento')
plot.show()
```

```python
tempos = dados.tempo.value_counts()
labels = dados.tempo.unique()

plot.pie(tempos, labels=labels, autopct='%1.1f%%')
plot.title('Tempos de tratamento')
plot.show()
```

```python
dados['tempo'].value_counts().plot.bar()
plot.title('Grafico de Barras: Tempo de tratamento')
plot.grid(True)
plot.xlabel("Horas")
plot.ylabel("Objetos de estudo")
plot.show()
```
